#### Representing Vectors

Now that you know how to represent vectors using their component and matrix form we can now hard-code them in Python. Let's say that you have the vectors:

$$ C = 1\hat{x} - 6\hat{y} \\
U = 8\hat{x} + 5\hat{y}\\
T = 7ax - 2ay + 3az \\
E = -9\hat{i} + 4\hat{j} + 8\hat{k}$$

In which it's matrix equivalent is:

$$ C = \begin{bmatrix} 1 \\ -6\end{bmatrix} , U = \begin{bmatrix}  8\\ 5\end{bmatrix} , T = \begin{bmatrix} 7 \\ -2 \\  3 \end{bmatrix}, E = \begin{bmatrix} -9 \\ 4\\ 8\end{bmatrix}
$$
$$ C = \begin{bmatrix} 1 & -6\end{bmatrix} , U = \begin{bmatrix} 8 & 5\end{bmatrix} , T = \begin{bmatrix} 7 & -2 & 3\end{bmatrix} , E = \begin{bmatrix} -9 & 4 & 8\end{bmatrix} 
$$

We can then start doing numpy code with this by:

```python
## Importing necessary libraries
import numpy as np ## 'np' here is short-hand name of the library (numpy) or a nickname.
```

```python
C = np.array([1, -6])
U = np.array([8, 5])
T = np.array([
              [7],
              [-2],
              [3]
])
E = np.array ([[-9],
              [4],
              [8]])
print('Vector C is ', C)
print('Vector U is ', U)
print('Vector T is ', T)
print('Vector E is ', E)
```

### MatPlotLib

MatPlotLib or MATLab Plotting library is Python's take on MATLabs plotting feature. MatPlotLib can be used vastly from graping values to visualizing several dimensions of data.

#### Visualizing Data

It's not enough just solving these vectors so might need to visualize them. So we'll use MatPlotLib for that. We'll need to import it first.

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
A = [1, -2]
B = [-7, 4]
plt.scatter(A[0], A[1], label='A', c='orange')
plt.scatter(B[0], B[1], label='B', c='turquoise')
plt.grid()
plt.legend()
plt.show()
```

```python
A = np.array([1, -1])
B = np.array([1, 5])
R = A + B
Magnitude = np.sqrt(np.sum(R**2))
plt.title("Resultant Vector\nMagnitude:{}" .format(Magnitude))
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.quiver(A[0], A[1], B[0], B[1], angles='xy', scale_units='xy', scale=1, color='green')
plt.quiver(0, 0, R[0], R[1], angles='xy', scale_units='xy', scale=1, color='black')
plt.grid()
plt.show()
print(R)
print(Magnitude)
Slope = R[1]/R[0]
print(Slope)
Angle = (np.arctan(Slope))*(180/np.pi)
print(Angle)
```

##Result

Plotting Three Vectors and show the Resultant Vector as a result.
Use Head to Tail Method.

```python
A = np.array([0, 5])
C = np.array([0, 2.5])
E = np.array([5, 5])
R = A + C + E
Magnitude = np.sqrt(np.sum(R**2))
plt.title("Resultant Vector\nMagnitude:{}" .format(Magnitude))
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.quiver(A[0], A[1], C[0], C[1], angles='xy', scale_units='xy', scale=1, color='green')
plt.quiver(0, 7.5, E[0], E[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, R[0], R[1], angles='xy', scale_units='xy', scale=1, color='black')
plt.grid()
plt.show()
print(R)
print(Magnitude)
Slope = R[1]/R[0]
print(Slope)
Angle = (np.arctan(Slope))*(180/np.pi)
print(Angle)
```
