###Transposition
Transposition, or the transpose of a matrix is done by interchanging the rows into column and vice versa. It is denoted by a superscript "T" in the given matrux [1]. Looking at the example below, we can see that the first column 4,6, and 9 became the first row in the transpose. 

$$
A=\begin{bmatrix} 4 & 8 & -2 \\ 6 & -3 & 5 \\ 9 & 3 & 7\end{bmatrix}\\
$$

$$
A^T=\begin{bmatrix} 4 & 6 & 9 \\ 8 & -3 & 3 \\ -2 & 5 & 7\end{bmatrix}\\
$$

In programming, transposition can be achieved using np.transpose() or using the T method as shown below

```python
G = np.array([
    [1,7,8],
    [3,9,0],
    [4,6,7]
])
```

```python
#np.transpose method
GT1 = np.transpose(G)
GT1
```

```python
#T method
GT2 = G.T
GT2
```

```python
#checking if the two methods would result in the same answer
np.array_equiv(GT1, GT2)
```

```python
#the shape of matrix that is not a square matrix also changes as it is transposed as shown below
R = np.array([
    [1,7,8,5],
    [3,9,0,4]
])
R.shape
```

```python
np.transpose(R).shape
```

```python
R.T.shape
```

###Dot Product/Inner Product
The dot product of two vectors refers to the sum of the products of elements with regards to the position. The first element of the first vector is multiplied by the first element of the second vector, and so on. The sum of these products is now the dot product [2]. In programming, this can be done using different functions as demonstrated below: 

```python
A = np.array([
    [4,-2,4],
    [3,9,1],
    [-1,2,8]
])
C = np.array([
    [2,7,0],
    [3,-9,8],
    [4,-5,1]
])
```

```python
#checking to see if the two matrices are equal
np.array_equiv(A, C)
```

```python
#np.dot() function
np.dot(A, C)
```

```python
#.dot() function
A.dot(C)
```

```python
#@ function
A @ C
```

```python
#np.matmul() function
np.matmul(A,C)
```

as shown above, there are multiple functions that can be used to perform the dot function which gives the same correct answer.

####Rule 1: 
The inner dimensions of the two matrices in question must be the same.

So given a matrix $E$ with a shape of $(a,b)$ where $a$ and $b$ are any integers. If we want to do a dot product between $E$ and another matrix $F$, then matrix $F$ should have a shape of $(b,c)$  where $b$ and $c$ are any integers. So for given the following matrices:

$$
E=\begin{bmatrix} 4 & -2 & 8 \\ 1 & 0 & 9 \\ -5 & 1 & 2\\ 3 & 9 & 6\end{bmatrix}\\
$$

$$
F=\begin{bmatrix} 2 & 7 & 5 \\ 1 & 3 & 0 \\ 5 & 8 & -2\\ 3 & -9 & 1\end{bmatrix}\\
$$

$$
G=\begin{bmatrix} 1 & 7 \\ 3 & 9 \\ 2 & 4 \end{bmatrix}\\
$$

So in this case $E$ has a shape of $(4,3)$, $F$ has a shape of $(4,3)$ and $G$ has a shape of $(3,2)$. So the only matrix pairs that is eligible to perform dot product is matrices $E·G$ or $F·G$.

```python
E = np.array([
    [4,-2,8],
    [1,0,9],
    [-5,1,2],
    [3,9,6]
])
F = np.array([
    [2,7,5],
    [1,3,0],
    [5,8,-2],
    [3,-9,1]
])
G = np.array([
    [1,7],
    [3,9],
    [2,4]
])
print(E.shape)
print(F.shape)
print(G.shape)
```

```python
E @ G
```

```python
F @ G
```

```python
#An error is displayed because it does not follow the rule
#Shape mismatch
E @ F
```

```python
#Taking the transpose of F changes it shape to (3,2), then dot multiplication is now possible
E @ F.T
```

####Rule 2: 
Dot Product has special properties


1.   $A·B$ $\neq$ $B·A$
2.   $A·(B·C)$ $=$ $(A·B)·C$
3.   $A·(B+C)$ $=$ $A·B+A·C$
4.   $(B+C)·A$ $=$ $B·A+C·A$
5.   $A·I$ $=$ $A$
6.   $A·∅$ $=$ $∅$


```python
A = np.array([
    [4,2,-6,5],
    [5,-1,7,9],
    [7,4,10,8],
    [-2,9,1,3]
])
B = np.array([
    [4,1,10,-7],
    [-8,2,4,1],
    [-5,2,9,-7],
    [1,3,6,2]
])
C = np.array([
    [1,-3,9,7],
    [0,-1,2,4],
    [6,5,2,-8],
    [1,-4,9,0]
])
```

the fifth property $A·I$ $=$ $A$ is demonstrated below:

```python
np.eye(4)
```

```python
D = A.dot(np.eye(4))
D
```

```python
np.array_equal(A, D)
```

as seen from the example above, matrix X when multiplied with an identity matrix of the same shape, would result to the a new matrix A with the same elements as matrix X. This could be checked manually or using the np.array_equal() function

###Determinant
The determinant can be viewed as a function whose input is a square matrix and whose output is a number. If n is the number of rows and columns in the matrix, the matrix can be called as an n×n matrix [3].

For a 2x2 matrix written as:
$$
\begin{bmatrix} a & b \\ c & d \end{bmatrix}\\
$$

then, the determinant can be calculated as:
$$
det(\begin{bmatrix} a & b \\ c & d \end{bmatrix}) = ad - bc
$$

higher square matrices are calculated in different ways which will be explained in future discussions.

in order to solve this in programming, the function np.linalg.det() is used.

```python
I = np.array([
    [4,2],
    [9,5]
])
np.linalg.det(I)
```

```python
S = np.array([
    [2,5,-6],
    [4,7,9],
    [-1,8,3]
])
np.linalg.det(S)
```

###Inverse
The inverse of a matrix is another fundamental operation in matrix algebra. Determining the inverse of a matrix let us determine if its solvability and its characteristic as a system of linear equation — we'll expand on this in the nect module. Another use of the inverse matrix is solving the problem of divisibility between matrices. Although element-wise division exists but dividing the entire concept of matrices does not exists. Inverse matrices provides a related operation that could have the same concept of "dividing" matrices.

Now to determine the inverse of a matrix we need to perform several steps. So let's say we have a matrix:
$$
M = \begin{bmatrix} 1 & 7 \\ -3 & 5 \end{bmatrix}\\
$$
 
First, we need to get the determinant of $M$.
$$
|M| = (1)(5) - (-3)(7) = 26
$$

Next, we need to reform the matrix into the inverse form:
$$
M^{-1}= {\frac{1}{|M|}}\begin{bmatrix} m_{(1,1)} & -m_{(0,1)} \\ -m_{(1,0)} & m_{(0,0)} \end{bmatrix}\\
$$
 
So that will be:
$$
M^{-1}= {\frac{1}{26}}\begin{bmatrix} 5 & -7 \\ 3 & 1 \end{bmatrix} = \begin{bmatrix} {\frac{5}{26}} & {\frac{7}{26}} \\ {\frac{3}{26}} & {\frac{1}{26}} \end{bmatrix}
$$

For higher-dimension matrices you might need to use co-factors, minors, adjugates, and other reduction techinques. To solve this programmatially we can use np.linalg.inv().

```python
B = np.array([
    [3,4],
    [-1, 7]
])

np.array(B @ np.linalg.inv(B), dtype=int)
```

```python
W = np.array([
              [3, 2, 5, 4],
              [4, 8, -2, 6],
              [4, 8, -2, -2],
              [9, -7, 3, 1]
])
T = np.linalg.inv(W)
T
```

```python
W @ T
```

**6 Matrix Multiplication Properties**

```python
#taking the matrices from the example under Rule 2:
A = np.array([
    [4,2,-6,5],
    [5,-1,7,9],
    [7,4,10,8],
    [-2,9,1,3]
])
B = np.array([
    [4,1,10,-7],
    [-8,2,4,1],
    [-5,2,9,-7],
    [1,3,6,2]
])
C = np.array([
    [1,-3,9,7],
    [0,-1,2,4],
    [6,5,2,-8],
    [1,-4,9,0]
])
```

**1st property**: $A·B$ $\neq$ $B·A$

```python
A @ B
```

```python
B @ A
```

```python
np.array_equal(A@B,B@A)
```

It can be seen that the results of the array as A is multiplied by B and B mutiplied by A are different, hence the False output proving that commutative property cannot be applied in dot multiplication of matrices.



---

**2nd Property:** $A·(B·C)$ $=$ $(A·B)·C$

```python
A@(B@C)
```

```python
(A@B)@C
```

```python
np.array_equal(A@(B@C),(A@B)@C)
```

The associative property of multiplication is demonstrated in this resulting to the same matrix proving that matrices can be multiplied in different groupings (parenthesis) but would result in an equal array. This is further proven using np.array_equal() function stating that the 2nd property is true.



---

**3rd Property:** $A·(B+C)$ $=$ $A·B+A·C$

```python
A@(B+C)
```

```python
A@B+A@C
```

```python
np.array_equal(A@(B+C),A@B+A@C)
```

As seen above, adding first the matrix B and C before multiplying with matrix A results to an equal array or answer when A is multiplied with B and C firsthand and then getting the sum of those two. This demonstrated the distributive property of multiplication. Another relationship is shown below in the fourth property which is essentially the same with the 3rd property. 



---

**4th Property:** $(B+C)·A$ $=$ $B·A+C·A$

```python
(B+C)@A
```

```python
B@A+C@A
```

```python
np.array_equal((B+C)@A,B@A+C@A)
```



---
